## Cleaning for Amazon_df

#### Split multi-values in "Category" feature to 3 Category ["" , "" , ""] 

```python
Amazon_df[["Category.1","Category.2","Category.3","Category.4","Category.5"]] = Amazon_df["category"].str.split("|",expand=True).iloc[:,:5]
```

```python
Amazon_df["Product_Name"] = Amazon_df["Category.4"] + " " + Amazon_df["Category.5"]
```

```python
Amazon_df["Product_Name"].fillna(Amazon_df["Category.4"],inplace=True)
```

#### Drop all duplicated Rows in "product_id"

```python
Amazon_df.duplicated().sum()
```

```python
Amazon_df = Amazon_df[~Amazon_df["product_id"].duplicated(keep=False)]
```

```python
Amazon_df.info()
```

#### Remove Features unnessasary

```python
Amazon_df.drop(columns=["Category.5","Category.4","product_name","category","about_product"],inplace=True)
```

```python
Amazon_df.head(5)
```

```python
Amazon_df[["product_id","user_name"]].head()
```

#### Remove special chararcter of these features ["discounted_price" , "actual_price"]

```python
Amazon_df["discounted_price"] = Amazon_df["discounted_price"].str.split("₹",expand=True)[1]
Amazon_df["actual_price"] = Amazon_df["actual_price"].str.split("₹",expand=True)[1]
```

#### Drop Sybol "%" From "discount_percentage" Column & ',' From [discounted_price , rating_count , actual_price] after that change Dtype to 'float'

```python
Amazon_df["discount_percentage"] = Amazon_df["discount_percentage"].str.split("%",expand=True)[0]
Amazon_df['actual_price'] = Amazon_df['actual_price'].str.replace(',', '').astype(float)
Amazon_df['discounted_price'] = Amazon_df['discounted_price'].str.replace(',', '').astype(float)
Amazon_df['discount_percentage'] = Amazon_df['discount_percentage'].astype(float)
Amazon_df['rating_count'] = Amazon_df['rating_count'].str.replace(',','').astype(int)
```

### Average prices of products per category

```python
df.groupby("Category.1")["discounted_price"].mean().round(2).plot(kind="bar",rot=90)
plt.title("Average prices of products per category")
plt.xlabel("Category")
plt.show()
```

## Highest and lowest product prices for the category

```python
MaxPrice_per_cate = df.groupby("Category.1")["discounted_price"].agg(["max","min"]).sort_values(by="Category.1").round(2)	
fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharex=True)

MaxPrice_per_cate["max"].plot(kind="bar",rot=90 , ax = axes[0],color="green")
MaxPrice_per_cate["min"].plot(kind="bar",rot=90 , ax = axes[1],color="red")
axes[0].set_title("Highest  product prices for the category")
axes[0].set_xlabel("Category")
axes[0].set_ylabel("price")
axes[1].set_title("lowest product prices for the category")
axes[1].set_xlabel("Category")
axes[1].set_ylabel("price")
plt.show()
# fig, ax = plt.subplots(figsize=(12, 6))
# MaxPrice_per_cate.plot(kind="bar", y=["max", "min"], color=["red", "green"], width=0.6, ax=ax)
# plt.show()
```

## Top 10 Product Name have Reviews and Least 10 

```python
Most_sum_reviews = df.groupby("Product_Name")["rating_count"].sum().sort_values(ascending=False).nlargest(10)
least_sum_reviews = df.groupby("Product_Name")["rating_count"].sum().sort_values(ascending=True).nsmallest(10)
fig, axes = plt.subplots(ncols=2, figsize=(16, 6), sharey=True)  

sns.barplot(x=Most_sum_reviews.values,y=Most_sum_reviews.index , ax= axes[0] , legend=False,orient="y",palette="Blues_r")
plt.xticks(rotation=90);
axes[0].set_title("Top 10 Product Name have most Reviews")
axes[0].set_xlabel("Review Count(Million)")
axes[0].set_ylabel("Product Name")

sns.barplot(x=least_sum_reviews.values,y=least_sum_reviews.index , ax= axes[1] , legend=False,orient="y",palette="Reds_r")
plt.xticks(rotation=90);
axes[1].set_title("Top 10 Product Name have less Reviews")
axes[1].set_xlabel("Review Count(thousands)")
plt.tight_layout()
plt.show();
```

```python
df.groupby("Category.1")["rating"].mean().sort_values(ascending=False).round(1)
```

```python
category_ratings = df.groupby("Category.1")["rating"].mean().sort_values(ascending=False).round(1)
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=category_ratings.values, y=category_ratings.index, palette="Blues_r")

for index, value in enumerate(category_ratings.values):
    ax.text(value + 0.05, index, f"{value} ★", fontsize=12, color="gold", fontweight="bold")

plt.xlabel("Average Rating", fontsize=12)
plt.ylabel("Category", fontsize=12)
plt.title("Average Rating per Category", fontsize=14, fontweight="bold")
plt.xlim(3.5, 5)  # ضبط نطاق المحور X ليتناسب مع القيم

plt.show()
```
